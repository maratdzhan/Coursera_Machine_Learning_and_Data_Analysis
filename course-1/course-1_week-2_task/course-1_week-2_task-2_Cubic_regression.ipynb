```python
import pandas as pd
import matplotlib.pyplot as plt
import scipy
import numpy as np
```

<h1 align = center>ЗАДАНИЕ 2: аппроксимация функции</h1>

<div><h3>Функция</h3></div>

```python
def f(x):
    return np.sin(x / 5) * np.exp(x / 10) + 5 * np.exp(-x / 2)
print('f(1)  = {:7.3f}; \nf(4)  = {:7.3f}; \nf(8)  = {:7.3f}; \nf(10) = {:7.3f}; \nf(15) = {:7.3f};\n'
      .format(f(1),f(4),f(8),f(10),f(15)))

osxi = np.arange(1,15.1,0.1)
```

```python
def pf(a,b,c,d,x):
    return (a*x**3 + b*x**2 + c*x+d);
```

<div>Матрица для многочлена первой степени</div>

```python
a1 = [[1,1],[1, 15]]
b1 = [[f(1)], [f(15)]]
f1 = np.linalg.solve(a1,b1)

f1 = f1[0] + f1[1]*osxi
plt.figure(figsize=(7,7))
fi = [f(i) for i in osxi]

plt.grid(True)
plt.plot(osxi,f1)
plt.plot(osxi,fi)
plt.show()
```

<div>Матрица для многочлена второй степени</div>

```python
a2 = [[1,1,1],[1,8,8**2],[1,15,225]]
b2 = [[f(1)],[f(8)], [f(15)]]
f2 = np.linalg.solve(a2,b2)
print(f2)
f2 = f2[0]+f2[1]*osxi+f2[2]*osxi**2
plt.figure(figsize=(7,7))
fi = [f(i) for i in osxi]
plt.grid(True)
plt.plot(osxi,f2)
plt.plot(osxi,fi)
plt.show()
```

```python
a3 = [[1.,1.,1.,1.],[1.,1.*4.,1.*4.**2,1.*(4.**3.)],[1.,1*10.,10.**2,1.*(10.**3.)],[1.,15.,15.**2.,15.**3.]]
# производная - это парабола: ax2+bx+c, где a = 1, b = -14, c = 40.
b3 = [[f(1.)],[f(4.)], [f(10.)], [f(15.)]]
f3 = np.linalg.solve(a3,b3)
print(f3)
f3 = f3[0]+f3[1]*osxi+f3[2]*(osxi**2)+f3[3]*(osxi**3)
plt.figure(figsize=(12,7))
fi = [f(i) for i in osxi]
plt.grid(True)
plt.plot(osxi,f3,'r-')
plt.plot(osxi,fi)
plt.show()
```
