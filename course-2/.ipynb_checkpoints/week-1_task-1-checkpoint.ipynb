```python
def make_bmi(height_inch, weight_pound):
    METER_TO_INCH, KILO_TO_POUND = 39.37, 2.20462
    return (weight_pound / KILO_TO_POUND) / \
           (height_inch / METER_TO_INCH) ** 2
```

```python
data['BMI'] = data.apply(lambda row: make_bmi(row['Height'], 
                                              row['Weight']), axis=1)
```

<br><br><div><h1 align = 'center'>Задание 2. Минимизация квадратичной ошибки</h1></div>

<br><br><div><h3 align = 'center'>6. Функция квадрата ошибки</h3></div>

```python
def Prediction(w0, w1, x):
    return (w0 + w1 * x);

def Error(w0, w1):
    return sum(map(lambda x, y: (y - Prediction(w0, w1, x)) ** 2, data.Weight, data.Height))
```

<br><br><div><h3 align = 'center'>7. Построение прямых с параметрами <i>w0,w1</i> = (60, 0.05) и (50, 0.16)</h3></div>

<div>Построение прямых с заранее известными плохими предсказаниями</div>

```python
data.plot.scatter(x='Weight', y='Height', title='Зависимость роста от веса', 
                  figsize = (11,8), alpha = 0.5, color = 'grey')
plt.plot(data.Weight, Prediction(60, 0.05, data.Weight), 
         color='red', linewidth=3, label = '60, 0.05', alpha = 0.7)
plt.plot(data.Weight, Prediction(50, 0.16, data.Weight), 
         color='blue', linewidth=3, label = '50, 0.16', alpha = 0.4)
plt.legend(fontsize = fontsize_value-5)
plt.grid(True)
plt.show()
```

<br><br><div><h3 align = 'center'>8. График зависимости функции ошибки от параметра  w1  при  w0  = 50</h3></div>

```python
plt.figure(figsize=(10,8))
plt.plot(np.linspace(-10, 10, 200), Error(50, np.linspace(-10, 10, 200)))
plt.grid(True)
plt.xlabel('w1')
plt.ylabel('error')
plt.title('Зависимость функции ошибки от w1')
plt.show()
```

<br><br><div><h3 align = 'center'>9. Поиск минимума функции ошибки</h3></div>

```python
from scipy.optimize import minimize_scalar 
```

```python
optimizing_result = minimize_scalar(lambda w1: Error(50, w1), bounds = (-5,5))
print(optimizing_result)
```

```python
data.plot.scatter(x='Weight', y='Height', title='Зависимость роста от веса', 
                  figsize = (11,8), alpha = 0.3, color = 'blue')
plt.plot(data.Weight, Prediction(60, 0.05, data.Weight), 
         color='green', linewidth=3, label = '60, 0.05', alpha = 0.4)
plt.plot(data.Weight, Prediction(50, 0.16, data.Weight), 
         color='black', linewidth=3, label = '50, 0.16', alpha = 0.4)
### Добавляем оптимальную величину
plt.plot(data.Weight, Prediction(50, optimizing_result.x, data.Weight), color='red', 
         linewidth=3, label = 'optimal (w1 = {:.4f})'.format(optimizing_result.x), alpha = 0.8)
###
plt.legend(fontsize = fontsize_value-5)
plt.grid(True)
plt.show()
```

<br><br><div><h3 align = 'center'>10. Построение 3D графиков</h3></div>

<br><br><div><h4 align = 'center'>10.1 Тренировочный график</h3></div>

<div>Функция $z(x,y) = sin(\sqrt{x^2+y^2})$ для значений $x$ и $y$ из интервала [-5,5] c шагом 0.25.</div>

```python
fig = plt.figure()
fig.set_size_inches(10, 7)
ax = fig.gca(projection='3d') # get current axis
# Создаем массивы NumPy с координатами точек по осям X и У. 
# Используем метод meshgrid, при котором по векторам координат 
# создается матрица координат. Задаем нужную функцию Z(x, y).
X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)
Z = np.sin(np.sqrt(X**2 + Y**2))

# Наконец, используем метод *plot_surface* объекта 
# типа Axes3DSubplot. Также подписываем оси.
surf = ax.plot_surface(X, Y, Z)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show()
```

<br><br><div><h4 align = 'center'>10.2 График исследуемой зависимости</h3></div>

```python
fig = plt.figure()
fig.set_size_inches(14, 10)
ax = fig.gca(projection='3d') # get current axis
# Создаем массивы NumPy с координатами точек по осям X и У. 
# Используем метод meshgrid, при котором по векторам координат 
# создается матрица координат. Задаем нужную функцию Z(x, y).
w0 = np.arange(-100, 100, 1)
w1 = np.arange(-5, 5, 0.25)
w0, w1 = np.meshgrid(w0, w1)
err = Error(w0,w1)

# Наконец, используем метод *plot_surface* объекта 
# типа Axes3DSubplot. Также подписываем оси.
surf = ax.plot_surface(w0, w1, err)
ax.set_xlabel('Intercept')
ax.set_ylabel('Slope')
ax.set_zlabel('Error')
plt.show()
```

<br><br><div><h3 align = 'center'>11. Минимизация функции ошибки с новыми параметрами</h3></div>

```python
from scipy import optimize
```

```python
result = optimize.minimize(lambda w: Error(w[0], w[1]), (0, 0),
                           bounds = ((-100,100), (-5,5)), 
                           method = 'L-BFGS-B')
print(result)
```

```python
data.plot.scatter(x='Weight', y='Height', title='Зависимость роста от веса', 
                  figsize = (11,8), alpha = 0.3, color = 'blue')
### Добавляем старую оптимальную величину
plt.plot(data.Weight, Prediction(50, optimizing_result.x, data.Weight), 
         color='green', linewidth=3, 
         label = 'old optimal w1 on [-5,5](x = {:.4f})'.format(optimizing_result.x), 
         alpha = 0.2)
###
### Добавляем новую оптимальную величину
plt.plot(data.Weight, Prediction(result.x[0], result.x[1], data.Weight), 
         color='red', linewidth=3, 
         label = 'optimal w0 on [-100, 100],\
 w1 on [-5,5] (w = ({:.3f}; {:.3f}))'.format(result.x[0], result.x[1]), 
         alpha = 0.75)
###
plt.legend(fontsize = fontsize_value-5)
plt.grid(True)
plt.show()
```
